### Initial Data Cleaning

```python
# Renaming a few columns for ease of use
df = df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
```

```python
# Cleaning data
df['rate'] = df['rate'].replace('NEW',np.NaN)
df['rate'] = df['rate'].replace('-',np.NaN)

# Changing categorical features for statstical computation
df.online_order = df.online_order.apply(lambda x: '1' if str(x)=='Yes' else '0')
df.book_table = df.book_table.apply(lambda x: '1' if str(x)=='Yes' else '0')


# Changing datatype of features for statstical computation
df.rate = df.rate.astype(str)
df.rate = df.rate.apply(lambda x : x.replace('/5',''))
df.rate = df.rate.astype(float)

df.cost = df.cost.astype(str)
df.cost = df.cost.apply(lambda y : y.replace(',',''))
df.cost = df.cost.astype(float)

# Viewing all null values across columns
df.isnull().sum()
```

### Exploring data to check the columns with the amount of missing data

```python
def missing_values(data, thresh = 25, color = 'black', edgecolor = 'black', height = 3, width = 15):
    
    plt.figure(figsize = (width, height))
    percentage = (data.isnull().mean()) * 100
    percentage.sort_values(ascending = False).plot.bar(color = color, edgecolor = edgecolor)
    plt.axhline(y = thresh, color = 'r', linestyle = '-')
    
    plt.title('Missing values percentage per column', fontsize = 20, weight = 'bold' )
    
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh + 12.5, f'Columns with more than {thresh}% missing values',
             fontsize = 12, color = 'crimson', ha = 'left' ,va = 'top')
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh - 5, f'Columns with less than {thresh}% missing values',
             fontsize=12, color='green', ha = 'left' ,va = 'top')
    plt.xlabel('Columns', size = 15, weight = 'bold')
    plt.ylabel('Missing values percentage')
    plt.yticks(weight = 'bold')
    
    return plt.show()

missing_values(data, thresh = 25, color = sns.color_palette('Reds',15))
```

As you can see from the graph that dish_liked has the most amount of missing values follwed by rate. We need to clean and format this data before we begin with Modeling

### Data Cleaning

```python
# Drop unwanted columns
df = df.drop(['url', 'address', 'phone', 'location'], axis = 1)
```

```python
# Dropping all rows with null values
df.dropna(how='any',inplace=True)
```

```python
# Removing duplicates
df.drop_duplicates(keep='first',inplace=True)
```

```python
#Converting caterogical variables into dummy variables
dummy_rest_type = pd.get_dummies(df['rest_type'])
dummy_type = pd.get_dummies(df['type'])
dummy_city = pd.get_dummies(df['city'])
dummy_cuisines = pd.get_dummies(df['cuisines'])
dummy_dishliked = pd.get_dummies(df['dish_liked'])
```

```python
# Concatenating dummy variables to the dataframe
df = pd.concat([df, dummy_rest_type, dummy_type, dummy_city, dummy_cuisines, dummy_dishliked],axis=1)
```

```python
# Viewing the data
df.head()
```

```python
# Deleting categorical variables since we added dummy columns
del df['rest_type']
del df['type']
del df['city']
del df['cuisines']
del df['dish_liked']
```

```python
# Viewing the data
df.head()
```

```python
# Dropping unwanted columns
df = df.drop(['name','reviews_list', 'menu_item'],axis=1)
```

```python
# Initializing Standard Scaler
scaler = StandardScaler()
```

```python
# Standardizing features by removing the mean and scaling to unit variance
x_fit = scaler.fit_transform(df)
```

```python
# Creating a new dataframe from the transformed data
X = pd.DataFrame(x_fit, columns=df.columns)
```

```python
# Viewing the new dataframe
X.head()
```

```python
# Dropping the target variable
X_init = X.drop(['rate'],axis=1)
```

```python
# Dropping all columns except target variable
Y_init = X.drop(X.columns.difference(['rate']),axis=1)
```

```python
# Initialzing the best feature selector 
bestfeatures = SelectKBest(k='all')
fit = bestfeatures.fit(X_init, Y_init)
```

```python
# Creating dataframe with scores from SelectKBest 
scores = pd.DataFrame(fit.scores_)
columns_ = pd.DataFrame(X_init.columns)
```

```python
# Concatenating columns and scores
featurescore = pd.concat([columns_, scores],axis=1)
```

```python
featurescore.columns = ['Features','Score']
```

```python
# Printing the first best 20 features based on the score
print(featurescore.nlargest(20,'Score'))
```

```python
# Selecting 800 best columns from 6990 columns
col_select = featurescore.nlargest(800, 'Score')
```

```python
# Dropping scores since it is not needed anymore
col_select.drop('Score', axis=1, inplace=True)
```

```python
col_select_list = list(col_select.Features)
```

```python
# Selecting 800 columns from the dataframe
x_select = X_init.loc[:,col_select_list]
```

```python
# Viewing the dataframe
x_select.head()
```

```python
# Eliminating duplicate columns
x_select = x_select.loc[:, ~x_select.columns.duplicated()]
```

```python
#Selecting target variable from the dataframe
Y_unscaled = df.rate
```

```python
# Splitting the dataframe into training and testing datasets
X_train, X_test, y_train, y_test=train_test_split(x_select, Y_unscaled, test_size = 0.05, random_state = 42)
```

```python
X_train
```

# Linear Regressor

```python
# Initializing and fitting Linear Regressor
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Prediction from Linear Regressor algorithm
y_pred_lr = lr.predict(X_test)

# R-squared value of Linear Regressor model
r2_score_lr = r2_score(y_test, y_pred_lr)

# Mean-squared value of Linear Regressor model
mse_score_lr = mean_squared_error(y_test, y_pred_lr)

print("R-2 score from the Linear Regressor is :", r2_score_lr, " Mean Squared Error is : ", mse_score_lr)
```

```python
# Predictions for the first 10 samples from the dataset
top_pred = lr.predict(X_test.iloc[0:10, :])
print("Predictions of the first 10 examples from the test dataset \n")
for idx, v in enumerate(top_pred):
    print("True Rating : ", y_test.iloc[idx], " and, Predicted Rating :", v)
```

# Decision Tree

```python
# Performing Grid Search and 5-fold Cross Validation for Decision Tree
dt_ds = GridSearchCV(estimator=DecisionTreeRegressor(), param_grid={"criterion": ["mse", "friedman_mse", "mae"], 
                                                                    "splitter": ["best"]})
dt_ds.fit(X_train, y_train)
```

```python
# Printing the results of Grid Search and Cross Valitdation 
dt_ds.cv_results_
```

```python
# Best parameters for the Decision Tree algorithm
dt_ds.best_params_
```

```python
# Predicting target variable using the best estimator/model returned by GridSearch CV
y_pred_dt = dt_ds.best_estimator_.predict(X_test.iloc[:, :])

# R-squared value of Decision Tree model
r2_score_dt = r2_score(y_test, y_pred_dt)

# R-squared value of Decision Tree model
mse_score_dt = mean_squared_error(y_test, y_pred_dt)

print("R-2 score from the Decision Tree is :", r2_score_dt, " Mean Squared Error is : ", mse_score_dt)
```

```python
# Predictions for the first 10 samples from the dataset
top_pred = dt_ds.best_estimator_.predict(X_test.iloc[0:10, :])
print("Predictions of the first 10 examples from the test dataset \n")
for idx, v in enumerate(top_pred):
    print("True Rating : ", y_test.iloc[idx], " and, Predicted Rating :", v)
```
